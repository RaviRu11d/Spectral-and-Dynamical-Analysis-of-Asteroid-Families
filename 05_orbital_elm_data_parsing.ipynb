```python
import os
import pathlib
import pandas as pd
```

```python
cols_prop = [
    "Asteroid_id",
    "H",
    "a_p",
    "e_p",
    "sin_i_p",
    "n",
    "g",
    "s",
    "LCE",
    "My"
]
```

```python
core_path = os.getcwd()
prop_df = pd.read_csv("data/lvl0/proper_elements.syn", delimiter='\s+', skiprows=2, names=cols_prop)
```

```python
cols_fam = [
    "Asteroid_id",
    "H",
    "status",
    "family_1",
    "dv_fam_1",
    "near_1",
    "family_2",
    "dv_fam_2",
    "near_2",
    "rescod"
]
```

```python
family_df = pd.read_csv("data/lvl0/family_classification.famrec", delimiter='\s+', names=cols_fam, comment='%')
```

```python
family_df.head()
```

```python
family_df.dtypes
```

```python
cols_rms = [
    "Asteroid_id",
    "rms_a",
    "rms_e",
    "rms_sin_i",
    "rms_n",
    "rms_g",
    "rms_s",
    "lam_fit_deg",
    "My"
]
```

```python
RMS_mb_df = pd.read_csv("data/lvl0/RMS_main_belt.sig", delimiter='\s+', names=cols_rms, comment='%')
```

```python
RMS_mb_df.head()
```

```python
RMS_mb_df.dtypes
```

```python
for df in [prop_df, family_df, RMS_mb_df]:
    df['Asteroid_id'] = df['Asteroid_id'].astype(str).str.strip()
```

```python
# Normalizing the 'Asteroid_id' column across all dataframes.

print("Prop: ", prop_df['Asteroid_id'].dtype, "-----", len(prop_df))
print("Family: ", family_df['Asteroid_id'].dtype, "-----", len(family_df))
print("RMS: ", RMS_mb_df['Asteroid_id'].dtype, "-----", len(RMS_mb_df))
```

```python
pathlib.Path(os.path.join(core_path, "data/lvl1")).mkdir(parents=True, exist_ok=True)

prop_df.to_pickle(os.path.join(core_path, "data/lvl1", "prop_df.pkl"), protocol=4)
family_df.to_pickle(os.path.join(core_path, "data/lvl1", "family_df.pkl"), protocol=4)
RMS_mb_df.to_pickle(os.path.join(core_path, "data/lvl1", "RMS_mb_df.pkl"), protocol=4)
```
