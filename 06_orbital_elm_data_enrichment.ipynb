```python
import os
import pathlib
import pandas as pd
import numpy as np
```

```python
core_path = os.getcwd()
prop_df = pd.read_pickle(os.path.join(core_path, "data/lvl1/", "prop_df.pkl"))
family_df = pd.read_pickle(os.path.join(core_path, "data/lvl1/", "family_df.pkl"))
RMS_mb_df = pd.read_pickle(os.path.join(core_path, "data/lvl1/", "RMS_mb_df.pkl"))
```

```python
prop_df.head()
```

```python
family_df.head()
```

```python
RMS_mb_df.head()
```

```python
prop_rms_df =  prop_df.copy()
prop_rms_df = prop_rms_df.merge(
    RMS_mb_df[['Asteroid_id', 'rms_a', 'rms_e', 'rms_sin_i', 'rms_n', 'rms_g', 'rms_s', 'lam_fit_deg']],
    on='Asteroid_id', how='left'
)

prop_rms_df.head()
```

```python
prop_family_df = prop_df.copy()
prop_family_df = prop_family_df.merge(
    family_df[['Asteroid_id', 'status', 'family_1', 'dv_fam_1', 'near_1', 'family_2', 'dv_fam_2', 'near_2', 'rescod']],
    on='Asteroid_id', how='left'
)

prop_family_df.head()
```

```python
orbital_df = prop_df.copy()
orbital_df = orbital_df.merge(
    RMS_mb_df[['Asteroid_id', 'rms_a', 'rms_e', 'rms_sin_i', 'rms_n', 'rms_g', 'rms_s', 'lam_fit_deg']],
    on='Asteroid_id', how='left'
)

orbital_df = orbital_df.merge(
    family_df[['Asteroid_id', 'status', 'family_1', 'dv_fam_1', 'family_2', 'dv_fam_2', 'rescod']],
    on='Asteroid_id', how='left'
)

orbital_df.head()
```

```python
print("=== Missing Values ===")
print(orbital_df.isnull().sum())
```

```python
print("\n=== Proper Element Ranges ===")
print(orbital_df[['a_p', 'e_p', 'sin_i_p', 'g', 's']].describe())
```

```python
# Check the outlier asteroid with max a_p
print(orbital_df.loc[orbital_df['a_p'].idxmax()])
```

```python
print("\n=== RMS Ranges ===")
print(orbital_df[['rms_a', 'rms_e', 'rms_sin_i', 'rms_g', 'rms_s']].describe())
```

```python
orbital_df['is_family_member'] = orbital_df['family_1'] != 0
# orbital_df_clean['is_family_member'] = orbital_df_clean['family_1'] != 0


print(f"Family members : {orbital_df['is_family_member'].sum():,}")
print(f"Background     : {(~orbital_df['is_family_member']).sum():,}")
print(f"Unique families (family_1): {orbital_df[orbital_df['is_family_member']]['family_1'].nunique()}")

print(f"\n=== Top 10 Families by Member Count ===")
print(
    orbital_df[orbital_df['is_family_member']]
    ['family_1']
    .value_counts()
    .head(10)
)
```

```python
print("=== LCE Distribution ===")
print(orbital_df['LCE'].describe())

# Negative LCE values in AstDyS mean the integration wasn't long enough — treat as unreliable
n_negative_lce = (orbital_df['LCE'] < 0).sum()
print(f"\nNegative LCE (unreliable integrations): {n_negative_lce:,}")
```

```python
before = len(orbital_df)

orbital_df_clean = orbital_df[
    (orbital_df['rms_a']    < 0.001) &
    (orbital_df['rms_e']    < 0.01)  &
    (orbital_df['rms_sin_i'] < 0.01)
].copy()

after   = len(orbital_df_clean)
removed = before - after

print(f"Before filtering : {before:,}")
print(f"After filtering  : {after:,}")
print(f"Removed (chaotic): {removed:,}  ({100*removed/before:.1f}%)")

# How many family members were lost
fam_before = orbital_df['is_family_member'].sum()
fam_after  = orbital_df_clean['is_family_member'].sum()
print(f"\nFamily members before : {fam_before:,}")
print(f"Family members after  : {fam_after:,}")
print(f"Family members lost   : {fam_before - fam_after:,}  ({100*(fam_before - fam_after)/fam_before:.1f}%)")
```

```python
# Main belt proper bounds
orbital_df = orbital_df[
    (orbital_df['a_p']     >= 2.0) & (orbital_df['a_p']     <= 3.5) &
    (orbital_df['e_p']     >= 0.0) & (orbital_df['e_p']     <= 0.4) &
    (orbital_df['sin_i_p'] >= 0.0) & (orbital_df['sin_i_p'] <= 0.6) &
    (orbital_df['g']       >= 0.0) & (orbital_df['g']       <= 150) &
    (orbital_df['s']       >= -120) & (orbital_df['s']       <= 0)
].copy()

print(f"After orbital bounds filter: {len(orbital_df):,}")
print(f"Removed: {1_092_819 - len(orbital_df):,}")
```

```python
orbital_df = orbital_df[orbital_df['LCE'] > 0].copy()
print(f"After LCE filter: {len(orbital_df):,}")
```

```python
for col in ['a_p', 'e_p', 'sin_i_p', 'g', 's', 'LCE']:
    idx = orbital_df[col].abs().idxmax()
    print(f"\n=== Max |{col}| ===")
    print(orbital_df.loc[idx, ['Asteroid_id', 'a_p', 'e_p', 'sin_i_p', 'g', 's', 'LCE', 'family_1']])
```

```python
for col in ['a_p', 'e_p', 'sin_i_p', 'g', 's', 'LCE']:
    print(f"{col:12s}  min: {orbital_df[col].min():.6f}   max: {orbital_df[col].max():.6f}")
```

```python
print("=== Final Dataset Summary ===")
print(f"Total asteroids  : {len(orbital_df):,}")
print(f"Family members   : {orbital_df['is_family_member'].sum():,}")
print(f"Background       : {(~orbital_df['is_family_member']).sum():,}")
print(f"Unique families  : {orbital_df[orbital_df['is_family_member']]['family_1'].nunique()}")
print(f"\nFamily ratio     : {100*orbital_df['is_family_member'].mean():.1f}%")
print(f"Background ratio : {100*(~orbital_df['is_family_member']).mean():.1f}%")
```

```python
pathlib.Path(os.path.join(core_path, "data/lvl2")).mkdir(parents=True, exist_ok=True)

prop_rms_df.to_pickle(os.path.join(core_path, "data/lvl2", "prop_rms_df.pkl"), protocol=4)
prop_family_df.to_pickle(os.path.join(core_path, "data/lvl2", "prop_family_df.pkl"), protocol=4)
orbital_df.to_pickle(os.path.join(core_path, "data/lvl2", "orbital_df.pkl"), protocol=4)
```
